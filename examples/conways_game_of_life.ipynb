# John Conway's Game Of Life: Threaded Edition

Some of the following code is adapted from https://jakevdp.github.io/blog/2013/08/07/conways-game-of-life/

```python
from time import sleep
from threading import Thread

import numpy as np

from ipycanvas import SketchyCanvas, hold_canvas
```

```python
def life_step(x):
    """Game of life step"""
    nbrs_count = sum(np.roll(np.roll(x, i, 0), j, 1)
                     for i in (-1, 0, 1) for j in (-1, 0, 1)
                     if (i != 0 or j != 0))
    return (nbrs_count == 3) | (x & (nbrs_count == 2))
```

```python
def draw(x, canvas, color='black'):
    with hold_canvas(canvas):
        canvas.clear()
        canvas.fill_style = '#FFF0C9'
        canvas.stroke_style = 'white'
        canvas.fill_rect(0, 0, canvas.size[0], canvas.size[1])

        canvas.fill_style = color
        canvas.stroke_style = color

        r = 0
        for row in x:
            c = 0
            for value in row:
                if value:
                    canvas.fill_rect(r * n_pixels, c * n_pixels, n_pixels, n_pixels)
                    canvas.stroke_rect(r * n_pixels, c * n_pixels, n_pixels, n_pixels)

                c += 1
            r += 1
```

```python
n_pixels = 15

canvas = SketchyCanvas(width=x.shape[1]*n_pixels, height=x.shape[0]*n_pixels)
canvas.fill_style = '#FFF0C9'
canvas.stroke_style = 'white'
canvas.fill_rect(0, 0, canvas.size[0], canvas.size[1])

canvas
```

```python
draw(x, canvas, '#5770B3')
```

```python
class GameOfLife(Thread):
    def __init__(self, x, canvas):
        self.x = x
        self.canvas = canvas
        super(GameOfLife, self).__init__()

    def run(self):
        for _ in range(1_000):
            self.x = life_step(self.x)
            draw(self.x, self.canvas, '#5770B3')

            sleep(0.1)

GameOfLife(x, canvas).start()
```
